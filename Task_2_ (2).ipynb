## Prompt Templates

Prompt templates make prompts reusable and dynamic instead of hardcoding values.

**ChatPromptTemplate**

```python
from langchain_core.prompts import ChatPromptTemplate

chat = ChatPromptTemplate.from_messages([
    ("system", "You are a teacher"),
    ("human", "Explain {topic}")
])

messages = chat.format_messages(topic="Neural Networks")

for msg in messages:
    print(msg)
```

**Input Validation**

```python
def validate(audience, tone):
    if audience not in ["beginner", "intermediate", "expert"]:
        audience = "beginner"
    if tone not in ["formal", "casual", "fun"]:
        tone = "formal"
    return audience, tone
```

**Prompt Generator**

```python
def generate_prompt(topic, audience, tone):
    audience, tone = validate(audience, tone)

    template = "Explain {topic} for {audience} in a {tone} tone."

    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template
    )

    return prompt.format(topic=topic, audience=audience, tone=tone)

print(generate_prompt("AI", "kids", "cool"))
```

**Reusability**

```python
cases = [
    {"topic": "AI", "audience": "beginner", "tone": "fun"},
    {"topic": "Python", "audience": "intermediate", "tone": "casual"},
]

for case in cases:
    print(prompt.format(**case))
```
